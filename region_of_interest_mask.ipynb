```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = plt.imread("test.jpg")
```

```python
print(type(image), image.shape)
```

```python
xsize = image.shape[1]
ysize = image.shape[0]
```

```python
region_select = np.copy(image)
```

```python
left_bottom = [0, 539]
right_bottom = [900, 300]
apex = [400, 0]
```

```python
fit_left = np.polyfit(
    (left_bottom[0], apex[0]), (left_bottom[1], apex[1]), 1
)

fit_right = np.polyfit(
    (right_bottom[0], apex[0]), (right_bottom[1], apex[1]), 1
)

fit_bottom = np.polyfit(
    (left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1
)
```

```python
np.polyfit( (370, 700), (0, 450), 1 )[1]
```

```python
print(fit_left)
print(fit_right)
print(fit_bottom)
```

```python
XX, YY = np.meshgrid(np.arange(0, xsize),
                     np.arange(0, ysize))
```

```python
## Find the regions inside the lines
```

```python
region_thresholds = (YY > (XX*fit_left[0] + fit_left[1])) & \
                    (YY > (XX*fit_right[0] + fit_right[1])) & \
                    (YY < (XX*fit_bottom[0] + fit_bottom[1]))
```

```python
region_select[region_thresholds] = [255, 0, 0]
```

```python
plt.imshow(region_select)
plt.show()
```
