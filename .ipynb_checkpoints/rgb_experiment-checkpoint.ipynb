```python
import random

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (20.0, 60.0)


fig, axes = plt.subplots(nrows=51, ncols=5, sharex=True)

rgb_val = 0

for row in range(0, 51):
    for col in range(0, 5):
        img = np.zeros([0,0,3],dtype=np.uint8)
        img_copy = np.copy(img)
        img_copy[:,:,0] = random.randint(0, 255)
        img_copy[:,:,1] = random.randint(0, 255)
        img_copy[:,:,2] = random.randint(0, 255)        
        plt.imshow(img_copy)
        rgb_val += 1
```
