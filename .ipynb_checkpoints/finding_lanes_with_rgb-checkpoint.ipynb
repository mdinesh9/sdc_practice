```python
plt.rcParams['figure.figsize'] = (20.0, 60.0)
fig, axes = plt.subplots(nrows=13, ncols=2, sharex=True)

thresh_index = 0

for row in range(0, 13):
    for col in range(0, 2):
        
        if thresh_index <= 24:
        
            rgb_threshold = [r_thresh[thresh_index], g_thresh[thresh_index], b_thresh[thresh_index]]
            threshold_filter = (image[:,:,0] < rgb_threshold[0]) \
                     | (image[:,:,1] < rgb_threshold[1]) \
                     | (image[:,:,2] < rgb_threshold[2])

            image_copy[threshold_filter] = [0, 0, 0]
            axes[row, col].imshow(image_copy)
            axes[row, col].set_title(rgb_threshold)

            thresh_index += 1
        else:
            axes[row, col].imshow(image_copy)        
```
